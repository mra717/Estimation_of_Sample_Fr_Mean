## Loading the Training Data 

```python
#Loading the Bernoulli training graphs 
x_train_bern = np.load('gdrive/MyDrive/xtrain_bern.npy')
y_train_bern = np.load('gdrive/MyDrive/ytrain_bern.npy')
```

```python
#Loading the Preffrential Attachment training graphs 
x_train_pref = np.load('gdrive/MyDrive/xtrain_pref.npy')
y_train_pref = np.load('gdrive/MyDrive/ytrain_pref.npy')
```

```python
#Loading the Stochastic Block training graphs 
x_train_sb = np.load('gdrive/MyDrive/xtrain_sb.npy')
y_train_sb = np.load('gdrive/MyDrive/ytrain_sb.npy')
```

```python
#Combing all Data Sets into one 
x_train_all = np.append(x_train_bern, x_train_pref, axis=0)
x_train_all = np.append(x_train_all, x_train_sb, axis=0)

y_train_all = np.append(y_train_bern, y_train_pref, axis=0)
y_train_all = np.append(y_train_all, y_train_sb, axis=0)
```

```python
print(len(y_train_all))

print(len(x_train_all))
```

### Preprocessing Data

Need to normalize the values in the sample grpahs. Becaus we know they have a maximum of 256 we divide by 255. Further our models require the inpput to be a tensor of shape (28,28,1) or (256,256,1)

```python
x_train_bern = x_train_bern/255
x_train_bern = np.expand_dims(x_train_bern, -1)
y_train_bern = np.expand_dims(y_train_bern, -1)

x_train_pref = x_train_pref/255
x_train_pref = np.expand_dims(x_train_pref, -1)
y_train_pref = np.expand_dims(y_train_pref, -1)

x_train_sb = x_train_sb/255
x_train_sb = np.expand_dims(x_train_sb, -1)
y_train_sb = np.expand_dims(y_train_sb, -1)

x_train_all = x_train_all/255
x_train_all = np.expand_dims(x_train_all, -1)
y_train_all = np.expand_dims(y_train_all, -1)
```

```python
y_train_all[0].shape
```

## Loading and Traing the General 28 node Unet mode (gen28model) 

```python
gen28model = Model.unet()
```

```python
gen28_hist = gen28model.fit(x=x_train_all, y=y_train_all, epochs=2000, validation_split=0.1)
```

```python
#saving the trained model 
gen28model.save("gen28model.h5")
```

```python
files.download("gen28model.h5")
```

```python
# convert the gen28_hist dict to a pandas DataFrame to save it as a csv file:     
gen28model_hist_df = pd.DataFrame(gen28_hist.history) 

gen28model_hist_df.to_csv('gen28model_hist.csv')
```
